```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL of the website
base_url = "https://carsheet.io/all-cars/2000-2025/sort-by-year-desc/"

# Initialize an empty list to store all car data
all_cars_data = []

# Initialize headers as None
headers = None

# Function to scrape data from a single page
def scrape_page(url):
    global headers  # Use the global headers variable
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table containing the car data
    table = soup.find('table', {'id': 'carsheet'})
    
    if table:
        # Extract headers only once (on the first page)
        if headers is None:
            headers = [header.find('div', {'class': 'th-wrap'}).text.strip() for header in table.find_all('th')]
        
        # Extract rows
        rows = table.find_all('tr')
        for row in rows[1:]:  # Skip the header row
            cells = row.find_all('td')
            car_data = [cell.text.strip() for cell in cells]
            all_cars_data.append(car_data)
    
    # Find the next page link
    next_page = soup.find('a', text='Next')
    if next_page:
        next_page_url = next_page['href']
        return base_url + next_page_url
    else:
        return None

# Start scraping from the first page
next_page_url = base_url
while next_page_url:
    print(f"Scraping page: {next_page_url}")
    next_page_url = scrape_page(next_page_url)

# Convert the scraped data to a DataFrame
if headers and all_cars_data:  # Ensure headers and data are available
    df = pd.DataFrame(all_cars_data, columns=headers)
    
    # Save the DataFrame to an Excel file
    df.to_excel('all_cars_data.xlsx', index=False)
    print("Scraping completed and data saved to 'all_cars_data.xlsx'")
else:
    print("No data found or headers not defined.")
```
